```python
import plotly.graph_objects as go

import numpy as np
np.random.randn(1)

x = np.random.randn(500)
y = np.random.randn(500) + 1
```

```python
fig = go.Figure(go.Histogram2d(
                x=x,
                y=y,
                ))
fig.show()
```

```python
fig = go.Figure(go.Heatmap(
        z = [[1, None, 30, 50], [10, 20, 30, 40], [12, 0, 5, 7]],
        x = ['Segunda', 'Terça', 'Quarta', 'Quinta'],
        y = ['Manha', 'Tarde', 'Noite']
))

fig.show()
```
